```python
from ultralytics import YOLO
import json
import cv2
import os

# Cargar el modelo YOLOv8
model = YOLO('/content/runs/detect/train/weights/best.pt')  # Reemplaza con la ruta a tu modelo

# Procesar imágenes y extraer detecciones
resultados_json = []
frames_dir = './video_frames'  # Directorio de los fotogramas extraídos
if not os.path.exists(frames_dir):
    os.makedirs(frames_dir)

for frame_file in os.listdir(frames_dir):
    if frame_file.endswith('.jpg'):
        path = os.path.join(frames_dir, frame_file)
        image = cv2.imread(path)

        # Realizar detección
        results = model(image)

        # Extraer detalles de las detecciones
        detalles_predicciones = []
        for result in results:
            # Verificar si hay detecciones
            if hasattr(result, 'pred') and len(result.pred) > 0:
                for i, (xmin, ymin, xmax, ymax, conf, cls) in enumerate(result.pred[0]):
                    detalles_predicciones.append({
                        "clase": result.names[int(cls)],
                        "bbox": [xmin.item(), ymin.item(), xmax.item(), ymax.item()],
                        "score": conf.item()
                    })

        resultados_json.append({
            "fotograma": frame_file,
            "predicciones": detalles_predicciones
        })

# Guardar en un archivo JSON
with open('detecciones_yolov8.json', 'w') as f:
    json.dump(resultados_json, f)
```

```python
import cv2
import os

def extraer_fotogramas(video_path, output_dir):
    if not os.path.exists(output_dir):
        os.makedirs(output_dir)

    cap = cv2.VideoCapture(video_path)
    frame_count = 0

    while True:
        ret, frame = cap.read()
        if not ret:
            break

        frame_file = os.path.join(output_dir, f"frame_{frame_count}.jpg")
        cv2.imwrite(frame_file, frame)
        frame_count += 1

    cap.release()
    print(f"Total de fotogramas extraídos: {frame_count}")

# Uso del script
video_path = "./vid.mp4"  # Ruta del video
extracted_frames = "./video_frames"  # Ruta de la carpeta de salida
extraer_fotogramas(video_path, extracted_frames)
```

```python
import json

# Cargar datos JSON de YOLOv8 y Detectron2
with open('detecciones_yolov8.json', 'r') as file:
    detecciones_yolov8 = json.load(file)

with open('detecciones_detectron2.json', 'r') as file:
    detecciones_detectron2 = json.load(file)

# Definir la clase objetivo para la comparación
clase_objetivo = 'Tiger'  # Cambia esto a la clase que deseas sumar

def suma_scores(detecciones, clase_objetivo):
    suma_scores = 0
    for deteccion in detecciones:
        for prediccion in deteccion.get('predicciones', []):
            if prediccion.get('clase') == clase_objetivo and prediccion.get('score') is not None:
                suma_scores += prediccion['score']
    return suma_scores

# Calcular la suma de scores para YOLOv8 y Detectron2
suma_scores_yolov8 = suma_scores(detecciones_yolov8, clase_objetivo)
suma_scores_detectron2 = suma_scores(detecciones_detectron2, clase_objetivo)

# Imprimir los resultados
print(f"Suma total de scores para '{clase_objetivo}' en YOLOv8: {suma_scores_yolov8}")
print(f"Suma total de scores para '{clase_objetivo}' en Detectron2: {suma_scores_detectron2}")

# Comparar cuál tiene mayor suma
if suma_scores_yolov8 > suma_scores_detectron2:
    print(f"YOLOv8 tiene una mayor suma de scores para la clase '{clase_objetivo}'.")
elif suma_scores_yolov8 < suma_scores_detectron2:
    print(f"Detectron2 tiene una mayor suma de scores para la clase '{clase_objetivo}'.")
else:
    print(f"Ambos modelos tienen sumas de scores iguales para la clase '{clase_objetivo}'.")
```

```python
def obtener_todas_las_clases(detecciones_json):
    todas_las_clases = set()
    for deteccion in detecciones_json:
        for prediccion in deteccion['predicciones']:
            todas_las_clases.add(prediccion['clase'])
    return todas_las_clases

# Cargar el archivo JSON de YOLOv8 y obtener todas las clases detectadas
with open('detecciones_yolov8.json', 'r') as file:
    detecciones_yolov8 = json.load(file)

todas_las_clases_yolov8 = obtener_todas_las_clases(detecciones_yolov8)

# Imprimir todas las clases detectadas por YOLOv8
print("Todas las clases detectadas por YOLOv8:", todas_las_clases_yolov8)
```
